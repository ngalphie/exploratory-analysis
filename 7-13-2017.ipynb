```python
import numpy as np
import pandas as pd

# Import pyplot for plotting
import matplotlib.pyplot as plt

# Some pretty Seaborn settings
import seaborn as sns
rc={'lines.linewidth': 2, 'axes.labelsize': 14, 'axes.titlesize': 14}
sns.set(rc=rc)

# Make Matplotlib plots appear inline
%matplotlib inline
```

```python
df = pd.read_csv('7-13-2017/mart-1-barcode12.csv', comment='#')
```

```python
df.head()
```

```python
df_have_cells = df.loc[df['Cell']==1,:]
```

```python
# Make the jitter plot
ax = sns.swarmplot(x='round', y='red', data=df, hue='Cell')
# Relabel axes
ax.set_xlabel('Round')
ax.set_ylabel('Cy5 Intensity (arb)');
fig = ax.get_figure()
fig.savefig("CY5.png")
```
